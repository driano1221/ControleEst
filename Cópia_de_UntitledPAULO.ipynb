```python
import pandas as pd
import numpy as np
grupos_analise = {
    'Feminino': {'variavel': 'F2', 'valor': 1, 'descricao': 'feminino'},
    'Masculino': {'variavel': 'F2', 'valor': 2, 'descricao': 'masculino'},
    'Centro_Oeste': {'variavel': 'F3', 'valor': 1, 'descricao': 'Centro-Oeste'},
    'Nordeste': {'variavel': 'F3', 'valor': 2, 'descricao': 'Nordeste'},
    'Norte': {'variavel': 'F3', 'valor': 3, 'descricao': 'Norte'},
    'Sudeste': {'variavel': 'F3', 'valor': 4, 'descricao': 'Sudeste'},
    'Sul': {'variavel': 'F3', 'valor': 5, 'descricao': 'Sul'},
    'ClasseA': {'variavel': 'P30', 'valor': 1, 'descricao': 'A'},
    'ClasseB': {'variavel': 'P30', 'valor': 2, 'descricao': 'B'},
    'ClasseC': {'variavel': 'P30', 'valor': 3, 'descricao': 'C'},
    'Idade_18_24': {'variavel': 'F6', 'valor': 1, 'descricao': '18 a 24 anos'},
    'Idade_25_34': {'variavel': 'F6', 'valor': 2, 'descricao': '25 a 34 anos'},
    'Idade_35_44': {'variavel': 'F6', 'valor': 3, 'descricao': '35 a 44 anos'},
    'Idade_45_64': {'variavel': 'F6', 'valor': 4, 'descricao': '45 a 64 anos'}
}
# Lista de variáveis para análise
variaveis = [
    'F2', 'F3', 'F6',
    'P0.1', 'P0.2', 'P0.3', 'P0.4', 'P0.5',
    'P1', 'P3', 'P5', 'P7',
    'P8_1', 'P8_2', 'P8_3', 'P8_4', 'P8_5', 'P8_6', 'P8_7', 'P8_8', 'P8_9',
    'P9_1', 'P9_2', 'P9_3', 'P9_4', 'P9_5', 'P9_6', 'P9_7',
    'P10', 'P11', 'P12', 'P13', 'P14', 'P15',
    'P17.1', 'P17.2', 'P17.3', 'P17.4', 'P17.5', 'P17.6', 'P17.7', 'P17.8', 'P17.9',
    'P18.1', 'P18.2','P30'
]

# Função para calcular estatísticas
def calcular_estatisticas(df, variaveis):
    resultados = []

    for var in variaveis:
        for onda in df['Onda'].unique():
            dados_onda = df[df['Onda'] == onda]

            dados_onda[var] = pd.to_numeric(dados_onda[var], errors='coerce')

            n = dados_onda[var].count()
            media = dados_onda[var].mean()
            desvio = dados_onda[var].std()
            erro_padrao = desvio / np.sqrt(n) if n > 0 else np.nan

            resultados.append({
                'Variavel': var,
                'Onda': f"{onda}ª - Geral",
                'N': n,
                'Media': media,
                'Desvio': desvio,
                'Erro_Padrao': erro_padrao
            })

    return pd.DataFrame(resultados)

# Calcular estatísticas
estatisticas = calcular_estatisticas(base_filtrada, variaveis)

# Exibir resultados
print(estatisticas)

# Salvar em Excel
# estatisticas.to_excel('estatisticas_por_ondag.xlsx', index=False)
```

```python
import pandas as pd
import numpy as np
from pandas import ExcelWriter

def calcular_estatisticas_ponderadas(df, variaveis, grupo_nome='Geral'):
    resultados = []

    for var in variaveis:
        for onda in df['Onda'].unique():
            dados_onda = df[df['Onda'] == onda].copy()

            # Converter para numérico e remover NAs
            dados_onda[var] = pd.to_numeric(dados_onda[var], errors='coerce')
            dados_onda['Peso'] = pd.to_numeric(dados_onda['Peso'], errors='coerce')

            # Remover linhas com NA na variável ou no peso
            validos = dados_onda[[var, 'Peso']].dropna()
            valores = validos[var]
            pesos = validos['Peso']

            n = len(validos)  # Número de observações válidas
            soma_pesos = pesos.sum()

            if n == 0 or soma_pesos == 0:
                media = desvio = erro_padrao = np.nan
            else:
                # Cálculos ponderados
                media = np.average(valores, weights=pesos)

                # Variância ponderada
                variancia = np.sum(pesos * (valores - media)**2) / soma_pesos
                desvio = np.sqrt(variancia)

                # Erro padrão considerando pesos
                erro_padrao = desvio / np.sqrt(soma_pesos)

            resultados.append({
                'Variável': var,
                'Grupo': grupo_nome,
                'Onda': f"Onda {int(onda)}",
                'N': n,
                'Soma_Pesos': soma_pesos,
                'Média': media,
                'Desvio_Padrão': desvio,
                'Erro_Padrão': erro_padrao
            })

    return pd.DataFrame(resultados)
```

```python
# Criar arquivo Excel com múltiplas abas
with ExcelWriter('Analise_Completa_Ponderada.xlsx') as writer:
    # 1. Análise Geral
    df_geral = calcular_estatisticas_ponderadas(base_filtrada, variaveis)
    df_geral.to_excel(writer, sheet_name='Geral', index=False)

    # 2. Análise por Grupos
    for grupo_nome, config in grupos_analise.items():
        # Filtrar o grupo
        grupo_df = base_filtrada[
            base_filtrada[config['variavel']] == config['valor']
        ].copy()

        # Calcular estatísticas para o grupo
        df_grupo = calcular_estatisticas_ponderadas(
            grupo_df,
            variaveis,
            grupo_nome=config['descricao']
        )

        # Salvar em aba separada
        df_grupo.to_excel(
            writer,
            sheet_name=config['descricao'][:31],  # Limite de caracteres do Excel
            index=False
        )

print("Análise completa exportada com sucesso!")
```
