### Introdução

No cenário atual, onde a **competitividade no mercado de e-commerce** cresce a passos largos, a capacidade de **otimizar investimentos em publicidade** se tornou um diferencial crucial para o sucesso das empresas. A área de Marketing da Melhores Compras, reconhecendo a importância estratégica de uma **alocação eficiente dos seus recursos**, buscou na ciência dos dados uma forma de **maximizar o retorno sobre investimento (ROI)** das suas campanhas publicitárias. Dentre as diversas técnicas disponíveis, a **regressão linear** emerge como uma ferramenta poderosa para estabelecer relações matemáticas entre os **gastos com publicidade** e o consequente **aumento nas vendas**.

Este estudo tem como objetivo **construir um modelo de regressão linear** que permita **prever o impacto das campanhas publicitárias futuras** nas vendas da empresa. Utilizando **dados históricos de gastos** em diversos canais de mídia, como **TV, Rádio, Redes Sociais**, entre outros, e os respectivos **incrementos de vendas**, buscamos identificar **quais mídias proporcionam o melhor retorno financeiro**. A partir dessa análise, a empresa poderá **otimizar sua estratégia de alocação de recursos**, direcionando investimentos para os canais mais rentáveis.

A **metodologia adotada** envolve a **transformação das variáveis categóricas** em variáveis dummy, uma **análise exploratória** para entender as correlações entre as variáveis, e a **construção de um modelo de regressão linear**. Além disso, será realizada uma **avaliação crítica dos resultados**, com o intuito de fornecer **recomendações claras e acionáveis** para a equipe de marketing. Espera-se, com isso, não apenas explicar os aumentos de vendas passados, mas também fornecer um **arcabouço robusto para a previsão de vendas futuras**, contribuindo para a **manutenção e expansão da competitividade** da Melhores Compras no mercado.

# Construção do Modelo de Regressão


```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# Variáveis independentes (gastos por mídia e as variáveis dummy)
X = dados_transformados.drop(columns=['Previsao Inicial de Aumento de Vendas (em mil unidades)'])
# Variável dependente (aumento nas vendas)
y = dados_transformados['Previsao Inicial de Aumento de Vendas (em mil unidades)']

# Dividindo os dados em conjuntos de treino e teste (70% treino, 30% teste)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Construção do Modelo de Regressão Linear
modelo = LinearRegression()
modelo.fit(X_train, y_train)

# Avaliação do modelo no conjunto de treino
y_train_pred = modelo.predict(X_train)
r2_treino = r2_score(y_train, y_train_pred)

# Avaliação do Modelo no Conjunto de Teste
y_test_pred = modelo.predict(X_test)
r2_teste = r2_score(y_test, y_test_pred)
mae = mean_absolute_error(y_test, y_test_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_test_pred))

# Coeficientes do modelo
coeficientes = pd.DataFrame({
    'Variável': X.columns,
    'Coeficiente': modelo.coef_
})

r2_treino, r2_teste, mae, rmse, coeficientes
```



#### Separação de Dados

Os dados foram divididos em variáveis independentes (X) e a variável dependente (y). As variáveis independentes incluem os gastos com publicidade por mídia, além das variáveis dummy que representam o tipo de mídia. A variável dependente é o aumento nas vendas.

A seguir, os dados foram divididos em conjuntos de treino e teste, utilizando 70% dos dados para treino e 30% para teste. Isso é importante para validar o desempenho do modelo em dados que não foram usados para treiná-lo.

#### Construção do Modelo de Regressão Linear

O modelo de regressão linear foi ajustado aos dados de treino. A métrica R², que mede a proporção da variabilidade dos dados explicada pelo modelo, foi calculada tanto para o conjunto de treino quanto para o conjunto de teste:

- **R² no Conjunto de Treino**: 0.869
- **R² no Conjunto de Teste**: 0.811


- Um valor de R² próximo de 1 indica que o modelo explica bem a variabilidade nos dados. Aqui, o modelo se saiu bem tanto no treino quanto no teste, sugerindo uma boa generalização.

#### Avaliação do Modelo no Conjunto de Teste

Além do R², outras métricas foram calculadas para avaliar o desempenho do modelo:

- **MAE (Erro Médio Absoluto)**: 316.18 unidades
- **RMSE (Raiz do Erro Quadrático Médio)**: 415.88 unidades


- O MAE e o RMSE indicam, em média, o quanto as previsões do modelo diferem dos valores reais. Um MAE de 316.18 significa que, em média, o modelo erra a previsão do aumento de vendas em 316.18 mil unidades. O RMSE, que penaliza erros maiores, é um pouco maior, indicando a presença de alguns erros maiores.

# Refinamento do Modelo


```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import Ridge, Lasso
from sklearn.model_selection import cross_val_score

# Criando termos polinomiais de grau 2 (quadrático)
poly = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly.fit_transform(X)

# Dividindo os dados em treino e teste novamente
X_poly_train, X_poly_test, y_poly_train, y_poly_test = train_test_split(X_poly, y, test_size=0.3, random_state=42)

# Ajustando o modelo polinomial
modelo_poly = LinearRegression()
modelo_poly.fit(X_poly_train, y_poly_train)

# Avaliando o modelo polinomial no conjunto de teste
y_poly_test_pred = modelo_poly.predict(X_poly_test)
r2_poly_teste = r2_score(y_poly_test, y_poly_test_pred)
mae_poly = mean_absolute_error(y_poly_test, y_poly_test_pred)
rmse_poly = np.sqrt(mean_squared_error(y_poly_test, y_poly_test_pred))

# Ajustando o modelo Ridge
ridge = Ridge(alpha=1.0)
ridge.fit(X_train, y_train)
ridge_scores = cross_val_score(ridge, X_train, y_train, cv=5, scoring='r2').mean()

# Ajustando o modelo Lasso
lasso = Lasso(alpha=0.1)
lasso.fit(X_train, y_train)
lasso_scores = cross_val_score(lasso, X_train, y_train, cv=5, scoring='r2').mean()

# Resultados
r2_poly_teste, mae_poly, rmse_poly, ridge_scores, lasso_scores
```



#### Regressões Polinomiais

Exploramos um modelo de regressão polinomial de grau 2 (quadrático) para capturar possíveis relações não lineares entre as variáveis. Após ajustar o modelo aos dados e avaliá-lo no conjunto de teste, os resultados foram:

- **R² no Conjunto de Teste (Modelo Polinomial)**: 0.954
- **MAE (Erro Médio Absoluto) do Modelo Polinomial**: 147.52 unidades
- **RMSE (Raiz do Erro Quadrático Médio) do Modelo Polinomial**: 204.17 unidades


- O modelo polinomial apresentou um R² muito alto no conjunto de teste (0.954), indicando que ele explica melhor a variabilidade dos dados em comparação com o modelo linear simples.
- O MAE e o RMSE são significativamente menores do que no modelo linear simples, sugerindo que o modelo polinomial está fazendo previsões mais precisas.

#### Regularização

Para evitar overfitting e melhorar a generalização, explorei os modelos de regularização Ridge e Lasso:

- **Ridge (R² Médio em Validação Cruzada)**: 0.804
- **Lasso (R² Médio em Validação Cruzada)**: 0.819


- A regularização Lasso apresentou um desempenho ligeiramente melhor do que Ridge, com um R² de 0.819 na validação cruzada.
- Esses modelos podem ser úteis quando queremos penalizar a complexidade do modelo, especialmente se estivermos preocupados com overfitting.

#### Considerações Finais

- **Modelo Polinomial**: Este modelo mostrou um desempenho excelente em termos de R², MAE e RMSE, sugerindo que ele captura bem as relações não lineares presentes nos dados.
- **Regularização**: Embora o modelo polinomial tenha mostrado um desempenho muito bom, a regularização (Ridge e Lasso) pode ser útil em cenários onde queremos simplificar o modelo e evitar a superestimação de determinadas variáveis.

Esses ajustes mostraram que, ao incluir termos polinomiais, o modelo pode capturar melhor as complexidades dos dados. A regularização serve como uma abordagem para melhorar a generalização do modelo, mantendo um bom desempenho preditivo.
