```python
from scipy import stats
import numpy as np

def realizar_testes(df, variaveis):
    resultados = []

    for var in variaveis:
        # Separar dados por onda
        grupo5 = df[df['Onda'] == 5][var].dropna()
        grupo6 = df[df['Onda'] == 6][var].dropna()
        grupo5 = pd.to_numeric(grupo5, errors='coerce')
        grupo6 = pd.to_numeric(grupo6, errors='coerce')
        # Teste de Levene para igualdade de variâncias
        try:  # Added try-except block
            levene_stat, levene_p = stats.levene(grupo5, grupo6)
        except ValueError:
            levene_stat, levene_p = np.nan, np.nan  # Assign NaN if Levene's test fails

        # Cálculos comuns
        n1, n2 = len(grupo5), len(grupo6)
        diff_media = grupo5.mean() - grupo6.mean()

        # Para variâncias iguais
        gl_equal = n1 + n2 - 2
        s_p = np.sqrt(((n1-1)*grupo5.var() + (n2-1)*grupo6.var()) / (n1+n2-2))
        se_diff_equal = s_p * np.sqrt(1/n1 + 1/n2)
        t_stat_equal = diff_media / se_diff_equal if se_diff_equal != 0 else np.nan  # Handle division by zero
        p_value_equal = 2 * (1 - stats.t.cdf(abs(t_stat_equal), gl_equal)) if not np.isnan(t_stat_equal) else np.nan # Handle NaN t-statistic

        # Para variâncias diferentes (Welch's t-test)
        # Check for conditions that lead to NaN before calculation
        if (grupo5.var()/n1)**2/(n1-1) + (grupo6.var()/n2)**2/(n2-1) == 0 or np.isnan((grupo5.var()/n1)**2/(n1-1) + (grupo6.var()/n2)**2/(n2-1)) :
            gl_unequal = np.nan #Assign NaN if denominator is zero or NaN
        else:
            gl_unequal = int(((grupo5.var()/n1 + grupo6.var()/n2)**2) /
                            ((grupo5.var()/n1)**2/(n1-1) + (grupo6.var()/n2)**2/(n2-1)))

        se_diff_unequal = np.sqrt(grupo5.var()/n1 + grupo6.var()/n2)
        t_stat_unequal = diff_media / se_diff_unequal if se_diff_unequal != 0 else np.nan  # Handle division by zero
        p_value_unequal = 2 * (1 - stats.t.cdf(abs(t_stat_unequal), gl_unequal)) if not np.isnan(t_stat_unequal) and not np.isnan(gl_unequal) else np.nan

        # Intervalos de confiança
        t_crit_equal = stats.t.ppf(0.975, gl_equal)
        t_crit_unequal = stats.t.ppf(0.975, gl_unequal)

        # Adicionar resultados para variâncias iguais
        resultados.append({
            'Variavel': var,
            'Tipo': 'Variâncias iguais assumidas',
            'Levene_Z': levene_stat,
            'Levene_Sig': levene_p,
            't': t_stat_equal,
            'df': gl_equal,
            'Sig_2_extremidades': p_value_equal,
            'Diferenca_media': diff_media,
            'Erro_padrao_diferenca': se_diff_equal,
            'IC_95_Inferior': diff_media - t_crit_equal * se_diff_equal,
            'IC_95_Superior': diff_media + t_crit_equal * se_diff_equal
        })

        # Adicionar resultados para variâncias diferentes
        resultados.append({
            'Variavel': var,
            'Tipo': 'Variâncias iguais não assumidas',
            'Levene_Z': None,
            'Levene_Sig': None,
            't': t_stat_unequal,
            'df': gl_unequal,
            'Sig_2_extremidades': p_value_unequal,
            'Diferenca_media': diff_media,
            'Erro_padrao_diferenca': se_diff_unequal,
            'IC_95_Inferior': diff_media - t_crit_unequal * se_diff_unequal,
            'IC_95_Superior': diff_media + t_crit_unequal * se_diff_unequal
        })

    return pd.DataFrame(resultados)

# Realizar os testes
resultados_testes = realizar_testes(base_filtrada, variaveis)

# Salvar em nova aba do mesmo arquivo Excel
with pd.ExcelWriter('estatisticas_por_ondag.xlsx', mode='a', engine='openpyxl') as writer:
    resultados_testes.to_excel(writer, sheet_name='Testes', index=False)

# Mostrar primeiros resultados
print(resultados_testes.head(10))
```

```python
# Filtrar apenas sexo masculino nas duas bases
base_masculino = base_filtrada[base_filtrada['F2'] == 2].copy()

# Função para estatísticas descritivas (mantém a mesma)
def calcular_estatisticas_masculino(df, variaveis):
    resultados = []

    for var in variaveis:
        for onda in df['Onda'].unique():
            dados_onda = df[df['Onda'] == onda]
            dados_onda[var] = pd.to_numeric(dados_onda[var], errors='coerce')
            n = dados_onda[var].count()
            media = dados_onda[var].mean()
            desvio = dados_onda[var].std()
            erro_padrao = desvio / np.sqrt(n) if n > 0 else np.nan

            resultados.append({
                'Variavel': var,
                'Onda': f"{onda}ª - masculino",
                'N': n,
                'Media': media,
                'Desvio': desvio,
                'Erro_Padrao': erro_padrao
            })

    return pd.DataFrame(resultados)

# Função modificada para testes
def realizar_testes_masculino(df, variaveis):
    resultados = []

    for var in variaveis:
        # Separar dados por onda
        grupo5 = df[df['Onda'] == 5][var].dropna()
        grupo6 = df[df['Onda'] == 6][var].dropna()
        grupo5 = pd.to_numeric(grupo5, errors='coerce')
        grupo6 = pd.to_numeric(grupo6, errors='coerce')
        try:
            # Teste de Levene para igualdade de variâncias
            levene_stat, levene_p = stats.levene(grupo5, grupo6)
        except:
            levene_stat, levene_p = np.nan, np.nan

        # Cálculos comuns
        n1, n2 = len(grupo5), len(grupo6)
        diff_media = grupo5.mean() - grupo6.mean()

        try:
            # Para variâncias iguais
            gl_equal = n1 + n2 - 2
            s_p = np.sqrt(((n1-1)*grupo5.var() + (n2-1)*grupo6.var()) / (n1+n2-2))
            se_diff_equal = s_p * np.sqrt(1/n1 + 1/n2)
            t_stat_equal = diff_media / se_diff_equal if se_diff_equal != 0 else 0
            p_value_equal = 2 * (1 - stats.t.cdf(abs(t_stat_equal), gl_equal))

            # Para variâncias diferentes (Welch's t-test)
            var1, var2 = grupo5.var(), grupo6.var()
            if var1 == 0 and var2 == 0:
                gl_unequal = gl_equal
                se_diff_unequal = se_diff_equal
                t_stat_unequal = t_stat_equal
                p_value_unequal = p_value_equal
            else:
                gl_unequal = int(((var1/n1 + var2/n2)**2) /
                                (((var1/n1)**2/(n1-1) if n1 > 1 else 0) +
                                 ((var2/n2)**2/(n2-1) if n2 > 1 else 0)))
                se_diff_unequal = np.sqrt(var1/n1 + var2/n2)
                t_stat_unequal = diff_media / se_diff_unequal if se_diff_unequal != 0 else 0
                p_value_unequal = 2 * (1 - stats.t.cdf(abs(t_stat_unequal), gl_unequal))

            # Intervalos de confiança
            t_crit_equal = stats.t.ppf(0.975, gl_equal)
            t_crit_unequal = stats.t.ppf(0.975, gl_unequal)

            # Calcular ICs
            ic_inf_equal = diff_media - t_crit_equal * se_diff_equal
            ic_sup_equal = diff_media + t_crit_equal * se_diff_equal
            ic_inf_unequal = diff_media - t_crit_unequal * se_diff_unequal
            ic_sup_unequal = diff_media + t_crit_unequal * se_diff_unequal

        except:
            # Se houver erro, preencher com NaN
            t_stat_equal = gl_equal = p_value_equal = se_diff_equal = ic_inf_equal = ic_sup_equal = np.nan
            t_stat_unequal = gl_unequal = p_value_unequal = se_diff_unequal = ic_inf_unequal = ic_sup_unequal = np.nan

        # Adicionar resultados para variâncias iguais
        resultados.append({
            'Variavel': var,
            'Tipo': 'Variâncias iguais assumidas',
            'Levene_Z': levene_stat,
            'Levene_Sig': levene_p,
            't': t_stat_equal,
            'df': gl_equal,
            'Sig_2_extremidades': p_value_equal,
            'Diferenca_media': diff_media,
            'Erro_padrao_diferenca': se_diff_equal,
            'IC_95_Inferior': ic_inf_equal,
            'IC_95_Superior': ic_sup_equal
        })

        # Adicionar resultados para variâncias diferentes
        resultados.append({
            'Variavel': var,
            'Tipo': 'Variâncias iguais não assumidas',
            'Levene_Z': None,
            'Levene_Sig': None,
            't': t_stat_unequal,
            'df': gl_unequal,
            'Sig_2_extremidades': p_value_unequal,
            'Diferenca_media': diff_media,
            'Erro_padrao_diferenca': se_diff_unequal,
            'IC_95_Inferior': ic_inf_unequal,
            'IC_95_Superior': ic_sup_unequal
        })

    return pd.DataFrame(resultados)

# Calcular estatísticas e testes
estatisticas_masculino = calcular_estatisticas_masculino(base_masculino, variaveis)
testes_masculino = realizar_testes_masculino(base_masculino, variaveis)

# Salvar em Excel
with pd.ExcelWriter('estatisticas_masculino.xlsx') as writer:
    estatisticas_masculino.to_excel(writer, sheet_name='Estatísticas', index=False)
    testes_masculino.to_excel(writer, sheet_name='Testes', index=False)

# Mostrar primeiros resultados
print("Estatísticas descritivas:")
print(estatisticas_masculino.head())
print("\nTestes:")
print(testes_masculino.head())
```
